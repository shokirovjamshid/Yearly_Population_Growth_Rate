```python
import matplotlib.pyplot as plt
import csv
plt.rcParams['axes.grid'] = True
```

```python
!wget -O Population.csv https://raw.githubusercontent.com/JavohirJalilov/AI-Matplotlib/master/dataset/Yearly%20Population%20Growth%20Rate.csv
```

```python
def read_data(filename):
  """
  Read csv file
  Args:
    file_path(str): file path
  Returns:
    list: data split row
  """
  # WRITE YOUR CODE HERE
  f = open(filename, mode='r+')
  data = list(csv.reader(f))
  return data
```

```python
data_list = read_data("Population.csv")
print(data_list)
```

```python
def get_data(data_list):
  """
  Get data from list.

  Args:
    data(list): data split row
  Returns:
    tuple: labels, points
  """
  # WRITE YOUR CODE HERE
  years = []
  population = []
  for i in data_list[1:]:
    years.append(int(i[0]))
    population.append(float(i[1]))
  return years,population
```

```python
labels, points = get_data(data_list)
```

```python
def plot_data(labels, points):
  """
  Show data.
  Args:
    labels(list): Years
    points(list): Population Growth
  Returns:
    None
  """
  # WRITE YOUR CODE HERE
  plt.plot(labels,points,color='orange')
  plt.xlabel('Years',color='g')
  plt.ylabel('Population Growth',color='g')
  plt.title('Yearly Population Growth Rate',color='b')
  plt.show()
```

```python
plt.figure(dpi=100)
plot_data(labels, points)
```
